# 01 - Data Exploration

Comprehensive exploratory data analysis of the DataPulse sample datasets.

**Datasets covered:**
- players.csv — 500 player records
- contests.csv — 1000 contests
- user_profiles.csv — 300 user profiles
- market_odds.csv — 1500 market records
- user_entries.csv — 2000 entry records

```python
import sys
sys.path.insert(0, '..')

import pandas as pd
import numpy as np
import plotly.express as px
from src.data.ingestion import DataLoader
from src.analysis.eda import AutoEDA
from src.visualization.plots import PlotFactory

print('Libraries loaded successfully')
```

```python
loader = DataLoader('../data')
players = loader.get_sample_data('players')
contests = loader.get_sample_data('contests')
users = loader.get_sample_data('user_profiles')
market = loader.get_sample_data('market_odds')
entries = loader.get_sample_data('user_entries')

print(f'Players: {players.shape}')
print(f'Contests: {contests.shape}')
print(f'Users: {users.shape}')
print(f'Market: {market.shape}')
print(f'Entries: {entries.shape}')
```

## Players Dataset Analysis

```python
eda = AutoEDA(players)
report = eda.run_full_analysis()

print(f"Shape: {report['overview']['shape']}")
print(f"Memory: {report['overview']['memory_usage_mb']:.2f} MB")
print(f"Numeric columns: {report['overview']['numeric_columns']}")
print(f"Duplicates: {report['overview']['duplicated_rows']}")

players.head(10)
```

```python
players.describe().round(2)
```

```python
fig = PlotFactory.distribution_plot(players['fantasy_points'], 'Fantasy Points Distribution')
fig.show()
```

```python
fig = PlotFactory.box_plot(players, 'position', 'fantasy_points', 'Fantasy Points by Position')
fig.show()
```

```python
fig = PlotFactory.correlation_heatmap(players, 'Player Stats Correlation Matrix')
fig.show()
```

```python
fig = PlotFactory.scatter_plot(players, 'salary', 'fantasy_points', color='position',
                               title='Salary vs Fantasy Points by Position')
fig.show()
```

## User Profiles Analysis

```python
user_eda = AutoEDA(users)
user_report = user_eda.run_full_analysis()

high_corr = user_report['correlations'].get('high_correlations', [])
print(f'High correlations (>0.7): {len(high_corr)}')
for item in high_corr[:5]:
    print(f"  {item['col_1']} <-> {item['col_2']}: {item['correlation']:.4f}")

users.describe().round(3)
```

```python
fig = PlotFactory.scatter_plot(users, 'win_rate', 'net_profit',
                               color='account_tier', title='Win Rate vs Net Profit by Account Tier')
fig.show()
```

```python
fig = PlotFactory.distribution_plot(users['win_rate'], 'User Win Rate Distribution')
fig.show()
```

## Market Odds Analysis

```python
market_eda = AutoEDA(market)
market_overview = market_eda.get_overview()
print(f"Shape: {market_overview['shape']}")
print(f"Columns: {market_overview['categorical_columns']}")

market['market_type'].value_counts()
```

```python
result_rate = market['result'].mean()
print(f'Overall win rate in market data: {result_rate:.1%}')
```

## Key Findings

1. **Fantasy Points:** Follow a roughly normal distribution centered around 35-40 points
2. **Salary-Performance Correlation:** Strong positive correlation between salary and fantasy points (r > 0.85)
3. **Position Differences:** Centers have highest rebounds, PGs have highest assists
4. **User Profitability:** Most users have negative net profit (expected house edge)
5. **Win Rate Distribution:** Beta distribution centered slightly below 50% for most users
6. **Market Balance:** Roughly 50/50 win/loss split in market results (random outcomes)
